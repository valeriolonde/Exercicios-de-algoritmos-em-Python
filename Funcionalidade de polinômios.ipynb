```python
class Polynomial(object):
    def __init__(self, coefficients):
        self.coeff = coefficients

    def __call__(self, x):
        """Avalia o polinômio."""
        s = 0
        for i in range(len(self.coeff)):
            s += self.coeff[i]*x**i
    return s

    def __add__(self, other):
        """Retorna a soma de dois objetos Polynomial."""
        # Dois casos:
        #
        # self: XXXXXXX
        # other: X X X
        #
        # ou:
        #
        # self: XXXXX
        # other: XXXXXXXX

        # Comece com a lista mais longa e adicione o outro
        if len(self.coeff) > len(other.coeff):
            result_coeff = self.coeff[:] # copie!
            for i in range(len(other.coeff)):
                result_coeff[i] += other.coeff[i]
        else:
            result_coeff = other.coeff[:] # copie!
            for i in range(len(self.coeff)):
                result_coeff[i] += self.coeff[i]

        return Polynomial(result_coeff)
    
    def __str__(self):
        terms = [f"{self.coeff[i]}*x^{i}" for i in range(len(self.coeff)) if self.coeff[i] != 0]
        if not terms:
            return "0"
        else:
            terms[0] = terms[0].lstrip('+')
return " + ".join(terms)PO
    

def test_Polynomial_str():
    ## Pergunta ao usuário quantos coeficientes ele gostaria de inserir:
    n = int(input("Quantos coeficientes deseja inserir? "))
    coefficients = []                                         ## os coeficientes são uma lista
    for i in range(n):                                        ## até range de n
        c = float(input(f"Digite o coeficiente de x^{i}: "))
        coefficients.append(c)

    # Cria um objeto Polynomial com a lista de coeficientes inserida
    p = Polynomial(coefficients)
        
    # verifica se a saída do método __str__() corresponde ao formato esperado
    expected_output = " + ".join([f"{coefficients[i]}*x^{i}" for i in range(n)])
    actual_output = str(p)       ## str vem da def__str__ 
    if actual_output == expected_output:
         print(f"Teste bem-sucedido! Resultado esperado: {expected_output}, \n Resultado atual: {actual_output}")
    else:
         print(f"Teste falhou. Resultado esperado: {expected_output}, Resultado atual: {actual_output}")

if __name__ == "__main__":
    test_Polynomial_str()


## Corrigindo o problema     

def test_Polynomial_str_2():
    ## Pergunta ao usuário quantos coeficientes ele gostaria de inserir:
    n = 0
    while n <= 0:
        try:
            n = int(input("Quantos coeficientes deseja inserir? "))
            if n <= 0:
                print("Por favor, digite um número inteiro positivo.")
        except ValueError:
            print("Por favor, digite um número inteiro positivo.")

    if n == 0:
        expected_output = ""
    else:
        coefficients = []                                         ## os coeficientes são uma lista
        for i in range(n):                                        ## até range de n
            c = float(input(f"Digite o coeficiente de x^{i}: "))
            coefficients.append(c)

        # Cria um objeto Polynomial com a lista de coeficientes inserida
        p = Polynomial(coefficients)

        # verifica se a saída do método __str__() corresponde ao formato esperado
        expected_output = " + ".join([f"{coefficients[i]}*x^{i}" for i in range(n)])
    actual_output = str(p)
    if actual_output == expected_output:
        print(f"Teste bem-sucedido! Resultado esperado: {expected_output}, Resultado atual: {actual_output}")
    else:
        print(f"Teste falhou. Resultado esperado: {expected_output}, Resultado atual: {actual_output}")

if __name__ == "__main__":
    test_Polynomial_str_2()

## utilizando numpy

import numpy as np

# pede ao usuário os coeficientes
n = int(input("Quantos coeficientes deseja inserir? "))
coefficients = []
for i in range(n):
    c = float(input(f"Digite o coeficiente de x^{i}: "))
    coefficients.append(c)

# cria um objeto polinomial com os coeficientes especificados
p = np.poly1d(coefficients)

# imprime a representação do polinômio
print(p)
```
